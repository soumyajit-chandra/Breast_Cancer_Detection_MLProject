```python
#slpitting data into two list
X=df.iloc[:,2:31].values #dependent
Y=df.iloc[:,1].values #independent
```

```python
#split data for train and test
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

```python
#Featurs scaling
from sklearn.preprocessing import StandardScaler
X_train=StandardScaler().fit_transform(X_train)
X_test=StandardScaler().fit_transform(X_test)
```

```python
#Model Algorithms
def models(X_train,Y_train):
    
    #Logistic Regrassion
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression(random_state=0)
    log.fit(X_train,Y_train)
    
    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    tree = DecisionTreeClassifier(random_state=0,criterion="entropy")
    tree.fit(X_train,Y_train)
    
    #Random forest
    from sklearn.ensemble import RandomForestClassifier
    forest=RandomForestClassifier(random_state=0,criterion="entropy",n_estimators=10)
    forest.fit(X_train,Y_train)
    
    print("1. Logistic Regrassion Accurecy:: ",log.score(X_train,Y_train)*100,"%")
    print("2. Decision Tree Accurecy:: ",tree.score(X_train,Y_train)*100,"%")
    print("3. Random forest Accurecy:: ",forest.score(X_train,Y_train)*100,"%")
    
    return log,tree,forest
```

```python
model=models(X_train,Y_train)
```

```python
#testing the Model

from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

for i in range(len(model)):
    print("Model ",i+1)
    print(classification_report(Y_test,model[i].predict(X_test)))
    print("Accuracy : ",accuracy_score(Y_test,model[i].predict(X_test)))
```

```python
#Prediction of random-Forest::
pred=model[2].predict(X_test)
print("Predicted Value")
print(pred)
print("Actual Value")
print(Y_test)
```

```python
from sklearn.metrics import confusion_matrix
cf_mat=confusion_matrix(Y_test,pred)
cf_mat
```

```python
tn,fp,fn,tp=cf_mat.ravel()
print(tn,fp,fn,tp)
```
